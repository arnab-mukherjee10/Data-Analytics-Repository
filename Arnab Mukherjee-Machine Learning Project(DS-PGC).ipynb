```python
# Replacing the elements of College column in ML case study data frame with numerical values.
# If college values matches with tier 1, it is replaced with value=3 as tier 1 has high weightage
# If college values matches with tier 2, it is replaced with value=2
# If college values matches with tier 3, it is replaced with value=1
```

```python
for i in df['College']:
    if i in Tier1:
        df['College'].replace(i, 3, inplace = True)
    elif i in Tier2:
        df['College'].replace(i, 2, inplace = True)
    elif i in Tier3:
        df['College'].replace(i, 1, inplace = True)
```

```python
# Replacing the elements of City column in ML case study data frame with numerical values.
# If city value with metro, it is assigned value=1
# If city value with non-metro, it is assigned value=0
```

```python
for i in df['City']:
    if i in metro:
        df['City'].replace(i, 1, inplace=True)
    elif i in non_metro:
        df['City'].replace(i, 0, inplace=True)
```

#### Preparing the data for test-train split

```python
# From the correlation matrix we saw that 'Graduation Marks' has a very low corrrelation coefficient with the target variable CTC.
# Since the value is very low(-0.005) as compared to other predictors, it means that there is almost no linear relationship between
# both of them. So we are dropping the 'Graduation Marks' from the training dataset to improve model accuracy.
```

#### Linear Regression using statsmodels

##### Steps required before training the Linear Regression model:-

##### 1.) Remove the outliers and impute the missing values if any.
##### 2.) Take all the predictor columns in one variable and the target/dependent variable in another column.
##### 3.) Split the dataset into train and test data in the ratio 80:20.
##### 4.) Using sn.add_constant() method add constant(B0) to the training dataset.

```python
import statsmodels.api as sn
```

```python
# Creating constant B0 for our model
```

```python
x_train_cons = sn.add_constant(x_train)
```

```python
x_train_cons.head()
```

```python
# Creating an object named lm of our OLS model and training the model with x_train and y_train values
```

```python
lm = sn.OLS(y_train, x_train_cons).fit()
```

```python
# Displaying the r2 score, f-statistic value, p-value  and other parameters of our model.
```

```python
lm.summary()
```

#### Preparing the data for Ridge and Lasso model

##### Steps required before training the Ridge and Lasso model:-

##### 1.) Remove the outliers and impute the missing values if any.
##### 2.) Take all the predictor columns in one variable and the target/dependent variable in another column.
##### 3.) Split the dataset into train and test data in the ratio 80:20.
##### 4.) Using StandardScaler() of preprocessing module create a scalar object.
##### 5.) Using the scalar object, transform(scale) the x_train and x_test data.

```python
from sklearn import preprocessing
```

```python
# Creating a scaler object using StandardScaler() method of preprocessing module
```

```python
scaler = preprocessing.StandardScaler().fit(x_train)
```

```python
# Scaling the x_train and x_test data using scaler object to form x_train_s and x_test_s scaled objects
```

```python
x_train_s = scaler.transform(x_train)
```

```python
x_test_s = scaler.transform(x_test)
```

#### Regression using Ridge model

```python
from sklearn.linear_model import Ridge
```

```python
# Creating an object named lm_r using Ridge Model
```

```python
lm_r = Ridge(alpha=0.5)
```

```python
# Training the model with alpha or lambda value=0.5
```

```python
lm_r.fit(x_train_s, y_train)
```

```python
r2_score(y_test, lm_r.predict(x_test_s))
```
